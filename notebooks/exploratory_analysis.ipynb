### Perguntas:

#### O que são esses valores negativos?

```python
colunas = [
    'VUF_QTPRODUTO',
    'VUF_VLRBRUTOVENDA',
    'VUF_VLRDESCONTO',
    'VUF_VLRLIQFINAL',
]

for coluna in colunas:
    valores_negativos = df[(df[coluna] < 0)]

    print('====='*23)
    print(f'Coluna - {coluna}:')
    display(valores_negativos.head(2))
```

#### De que forma esses valores podem influência em nossa análise?

```python
for coluna in colunas:
    valores_negativos = df[(df[coluna] < 0)]

    descrição = {
        'qtd_negativos': valores_negativos.shape[0],
        'valor_total': valores_negativos[[coluna]].sum(),
        'maior_valor': valores_negativos[[coluna]].min()
    }
    print('=='*20)
    df_descrição = pd.DataFrame(descrição)
    display(df_descrição.T)
```

#### Qual é o número de `Produtos` por `Boleto`?

```python
(
    df.rename(columns={'VUF_QTBOLETO':'qtd_produtos', 'VUF_CODIGO_BOLETO':'codigo_boleto'})
    .groupby('codigo_boleto')
    .agg({'qtd_produtos':'sum'})
    .reset_index()
)
```

**Respostas:** *Aqui é apresentada a quantidade de produtos distintos comprados no mesmo boleto. Não se engane, o que vemos na coluna 'qtd_produtos' não representa a quantidade de unidades de um mesmo produto, mas sim a quantidade de produtos distintos comprados no mesmo boleto.*

#### Quais são os clientes?

```python
(
    df.rename(columns={'CLV_BANCO':'clientes', 'VUF_CODIGO':'qtd_pedidos'})
    .groupby('clientes')
    .agg({'qtd_pedidos':'count'})
    .reset_index()
)
```
